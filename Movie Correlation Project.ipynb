```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None


# Now we need to read in the data
df = pd.read_csv(r'C:\Users\Vasso\Desktop\Portfolio projects\movies.csv')
```

```python
df
```

```python
# check for missing data
df = df.dropna()
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
 
```

```python
# check the Data Types for our columns

print(df.dtypes)
```

```python
# Create yaercorrect column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

# change data types of columns

df['gross']=df['gross'].astype('int64')

df['budget']=df['budget'].astype('int64')
```

```python
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
# Correlations

# Scatter plot budget

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross earnings')
plt.xlabel('Gross earnings')
plt.ylabel('Budget for film')

plt
plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x="gross", y="budget", data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Find Outliers

df.boxplot(column=['gross'])
```

```python
df.corr() #pearson
```

```python
# Correlation Matrix between all numeric columns

df.corr(method ='pearson')

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Turn columns that are type "object" into numeric values, in order to include them in the correlation matrix

df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == "object"):
        df_numerized[col_name] = df_numerized[col_name].astype("category")
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized

# Using factorize - this assigns a random numeric value for each unique categorical value  --- 2nd way

# df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```
